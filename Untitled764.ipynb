```python
import pandas as pd
import xarray as xr
import numpy as np
from scipy.spatial import cKDTree

# Φόρτωση αρχείων
csv_file_path = "yearly_SURF_ppb_O3.csv"
netcdf_file_path = "BaseCase_PERT_O3_YEARLY.nc"

df_csv = pd.read_csv(csv_file_path)
ds_netcdf = xr.open_dataset(netcdf_file_path)

# Εξαγωγή συντεταγμένων από το NetCDF
lon_grid = ds_netcdf["lon"].values
lat_grid = ds_netcdf["lat"].values

# Δημιουργία πλέγματος συντεταγμένων (grid)
lon_mesh, lat_mesh = np.meshgrid(lon_grid, lat_grid)
grid_points = np.column_stack([lon_mesh.ravel(), lat_mesh.ravel()])

# Δημιουργία KDTree για γρήγορη εύρεση πλησιέστερου γείτονα
tree = cKDTree(grid_points)

# Συντεταγμένες των σταθμών από το CSV
station_points = df_csv[['lon', 'lat']].values

# Εύρεση του πλησιέστερου κελιού του πλέγματος για κάθε σταθμό
_, indices = tree.query(station_points)

df_csv['matched_lon'] = grid_points[indices, 0]
df_csv['matched_lat'] = grid_points[indices, 1]

# Εξαγωγή δεδομένων όζοντος από το NetCDF
surf_ppb_o3_grid = ds_netcdf["SURF_ppb_O3"].values[0, :, :]

grid_o3_values = []
for lon, lat in zip(df_csv['matched_lon'], df_csv['matched_lat']):
    lon_idx = np.argmin(np.abs(lon_grid - lon))
    lat_idx = np.argmin(np.abs(lat_grid - lat))
    grid_o3_values.append(surf_ppb_o3_grid[lat_idx, lon_idx])

df_csv['grid_SURF_ppb_O3'] = grid_o3_values

# Αποθήκευση του αποτελέσματος
output_file = "matched_stations_with_grid_O3.csv"
df_csv.to_csv(output_file, index=False)

print(f"Αποθηκεύτηκε το αρχείο: {output_file}")
```

```python
import pandas as pd
import xarray as xr
import numpy as np
from sklearn.neighbors import BallTree
from geopy.distance import great_circle

# Φόρτωση αρχείων
csv_file_path = "yearly_SURF_ppb_O3.csv"
netcdf_file_path = "BaseCase_PERT_O3_YEARLY.nc"

df_csv = pd.read_csv(csv_file_path)
ds_netcdf = xr.open_dataset(netcdf_file_path)

# Εξαγωγή συντεταγμένων από το NetCDF
lon_grid = ds_netcdf["lon"].values
lat_grid = ds_netcdf["lat"].values

# Δημιουργία λίστας συντεταγμένων του πλέγματος (σε μοίρες)
grid_points = np.column_stack([lat_grid.repeat(len(lon_grid)), np.tile(lon_grid, len(lat_grid))])

# Μετατροπή σε ακτίνια για χρήση με BallTree
grid_points_rad = np.radians(grid_points)

# Δημιουργία BallTree για γρήγορη εύρεση πλησιέστερου γείτονα με Haversine απόσταση
tree = BallTree(grid_points_rad, metric='haversine')

# Συντεταγμένες των σταθμών από το CSV (σε μοίρες)
stations = df_csv[['lat', 'lon']].values

# Μετατροπή των σταθμών σε ακτίνια
stations_rad = np.radians(stations)

# Εύρεση του πλησιέστερου κελιού του πλέγματος για κάθε σταθμό
_, indices = tree.query(stations_rad, k=1)

df_csv['matched_lat'] = grid_points[indices.flatten(), 0]
df_csv['matched_lon'] = grid_points[indices.flatten(), 1]

# Εξαγωγή δεδομένων όζοντος από το NetCDF
surf_ppb_o3_grid = ds_netcdf["SURF_ppb_O3"].values[0, :, :]

grid_o3_values = []
for lat, lon in zip(df_csv['matched_lat'], df_csv['matched_lon']):
    lon_idx = np.argmin(np.abs(lon_grid - lon))
    lat_idx = np.argmin(np.abs(lat_grid - lat))
    grid_o3_values.append(surf_ppb_o3_grid[lat_idx, lon_idx])

df_csv['grid_SURF_ppb_O3'] = grid_o3_values

# Αποθήκευση του αποτελέσματος
output_file = "matched_stations_with_grid_O3-c.csv"
df_csv.to_csv(output_file, index=False)

print(f"Αποθηκεύτηκε το αρχείο: {output_file}")
```
