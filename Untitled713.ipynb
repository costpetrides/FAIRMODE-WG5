```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import plotly
import plotly.express as px
import plotly.graph_objects as go
import cufflinks as cf

import warnings
warnings.filterwarnings("ignore")
warnings.warn("this will not show")

plt.rcParams["figure.figsize"] = (10,6)

sns.set_style("whitegrid")
pd.set_option('display.float_format', lambda x: '%.3f' % x)

pd.set_option('display.max_columns', None)


# Read the three datasets
YO3 = pd.read_csv("/content/yearly_SURF_ppb_O3.csv")
YNO2 = pd.read_csv("/content/yearly_SURF_ug_NO2.csv")
YPM = pd.read_csv("/content/yearly_SURF_ug_PM25_rh50.csv")

# Merge the datasets on 'longitude' and 'latitude'
df = YO3.merge(YNO2, on=['lon', 'lat'], how='inner')\
               .merge(YPM, on=['lon', 'lat'], how='inner')

# Save the merged dataset
df.to_csv("merged_data.csv", index=False)

print("Merged dataset saved as 'merged_data.csv'")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numeric columns and reverse the order
columns = df.select_dtypes(include=['number']).columns.tolist()[::-1]  # Reverse order

# Define subplot grid size
num_columns = 3  # Number of columns in subplot grid
num_rows = (len(columns) + num_columns - 1) // num_columns  # Calculate rows dynamically

# Create subplots
fig, axes = plt.subplots(num_rows, num_columns, figsize=(15, 5 * num_rows))

# Flatten axes for iteration
axes = axes.flatten()

# Plot histograms with KDE for each column in reversed order
for ax, column in zip(axes, columns):
    sns.histplot(data=df, x=column, ax=ax, kde=True, bins=30)  # Adjust bins for clarity
    ax.set_title(column, fontsize=12)
    ax.set_xlabel(column, fontsize=10)
    ax.set_ylabel('Frequency', fontsize=10)

# Hide unused subplot axes
for i in range(len(columns), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()  # Adjust layout for better readability
plt.show()
```

## O3

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import plotly.graph_objects as go
import plotly.subplots as sp

# Perform seasonal decomposition
result = seasonal_decompose(df['SURF_ppb_O3'], model='additive', period=365)

# Create subplots for the decomposed components
fig = sp.make_subplots(
    rows=4, cols=1, shared_xaxes=True,
    subplot_titles=['Observed', 'Trend', 'Seasonal', 'Residual']
)

# Add traces for each component
fig.add_trace(go.Scatter(x=df.index, y=result.observed, mode='lines', name='Observed'), row=1, col=1)
fig.add_trace(go.Scatter(x=df.index, y=result.trend, mode='lines', name='Trend'), row=2, col=1)
fig.add_trace(go.Scatter(x=df.index, y=result.seasonal, mode='lines', name='Seasonal'), row=3, col=1)
fig.add_trace(go.Scatter(x=df.index, y=result.resid, mode='lines', name='Residual'), row=4, col=1)

# Update layout with the correct template
fig.update_layout(
    template='plotly',
    height=800,
    title='O3',
    title_x=0.5
)

# Show the figure
fig.show()
```

```python
import netCDF4
import numpy as np
import matplotlib.pyplot as plt

# Open the NetCDF file
O3map = netCDF4.Dataset('BaseCase_PERT_O3_YEARLY.nc', 'r')

# Extract variables: lon, lat, and O3 concentration
lon = O3map.variables['lon'][:]
lat = O3map.variables['lat'][:]
O3 = O3map.variables['SURF_ppb_O3'][0, :, :]  # Only first time point

# Close the dataset
O3map.close()

# Mask out invalid data (_FillValue)
fill_value = 9.969209968386869e+36
O3 = np.ma.masked_equal(O3, fill_value)

# Create a 2D meshgrid for lon & lat
Lon, Lat = np.meshgrid(lon, lat)

# Plot the O3 concentration
plt.figure(figsize=(10, 6))
plt.pcolormesh(Lon, Lat, O3, cmap='viridis', shading='auto')  # Use pcolormesh for better scaling
plt.colorbar(label='O3 Concentration (ppb)')
plt.title('Ozone (O3) Concentration Across the Region')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Create a figure with a map projection
fig, ax = plt.subplots(figsize=(10, 6), subplot_kw={'projection': ccrs.PlateCarree()})

# Plot the O₃ concentration heatmap
mesh = ax.pcolormesh(Lon, Lat, O3, cmap='viridis', shading='auto', transform=ccrs.PlateCarree())

# Add a colorbar
plt.colorbar(mesh, label='O₃ Concentration (ppb)')

# Add borders for better context
ax.add_feature(cfeature.BORDERS, linewidth=1, edgecolor="black")  # Country borders
ax.add_feature(cfeature.COASTLINE, linewidth=1, edgecolor="black")  # Coastline
ax.add_feature(cfeature.STATES, linewidth=0.5, edgecolor="black")  # State borders

# Title and labels
plt.title("Ozone (O₃) Concentration with Borders")
plt.xlabel("Longitude")
plt.ylabel("Latitude")

# Show the plot
plt.show()
```

```python
import netCDF4
import numpy as np
import pandas as pd  # Ensure pandas is imported

# Open the NetCDF file for NO2
NO2map = netCDF4.Dataset('BaseCase_PERT_NO2_YEARLY.nc', 'r')  # Replace with your actual file name

# Extract the variables
lon = NO2map.variables['lon'][:]
lat = NO2map.variables['lat'][:]
time = NO2map.variables['time'][:]
NO2 = NO2map.variables['SURF_ug_NO2'][:]  # Shape: (time, lat, lon)

# Close the dataset after reading the necessary variables
NO2map.close()

# Mask out invalid data (_FillValue)
fill_value = 9.969209968386869e+36  # The FillValue for invalid data
NO2 = np.ma.masked_equal(NO2, fill_value)

# Flatten the 3D NO2 data (time, lat, lon) into 2D for the first time point
# Assuming there's only one time point here
time = np.repeat(time, lon.size * lat.size)  # Repeat the time for each lat/lon combination
lat_lon_pairs = np.array(np.meshgrid(lat, lon)).T.reshape(-1, 2)  # Create all combinations of lat/lon

# Create a DataFrame
NO2map_df = pd.DataFrame(lat_lon_pairs, columns=['Latitude', 'Longitude'])
NO2map_df['Time'] = time
NO2map_df['NO2 Concentration (µg/m³)'] = NO2[0, :, :].flatten()  # Flatten the NO2 array for the first time point

# Remove rows with invalid NO2 concentration (masked)
NO2map_df = NO2map_df[NO2map_df['NO2 Concentration (µg/m³)'].notna()]

# Save to CSV
NO2map_df.to_csv('NO2_BaseCase.csv', index=False)

print("CSV file saved successfully!")
```
