# Data Imports !

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import plotly
import plotly.express as px
import plotly.graph_objects as go
import cufflinks as cf

import warnings
warnings.filterwarnings("ignore")
warnings.warn("this will not show")

plt.rcParams["figure.figsize"] = (10,6)

sns.set_style("whitegrid")
pd.set_option('display.float_format', lambda x: '%.3f' % x)

pd.set_option('display.max_columns', None)


# Read the three datasets
df1 = pd.read_csv("/content/yearly_SURF_ppb_O3.csv")
df2 = pd.read_csv("/content/yearly_SURF_ug_NO2.csv")
df3 = pd.read_csv("/content/yearly_SURF_ug_PM25_rh50.csv")

# Merge the datasets on 'longitude' and 'latitude'
df = df1.merge(df2, on=['lon', 'lat'], how='inner')\
               .merge(df3, on=['lon', 'lat'], how='inner')

# Save the merged dataset
df.to_csv("merged_data.csv", index=False)

print("Merged dataset saved as 'merged_data.csv'")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numeric columns and reverse the order
columns = df.select_dtypes(include=['number']).columns.tolist()[::-1]  # Reverse order

# Define subplot grid size
num_columns = 3  # Number of columns in subplot grid
num_rows = (len(columns) + num_columns - 1) // num_columns  # Calculate rows dynamically

# Create subplots
fig, axes = plt.subplots(num_rows, num_columns, figsize=(15, 5 * num_rows))

# Flatten axes for iteration
axes = axes.flatten()

# Plot histograms with KDE for each column in reversed order
for ax, column in zip(axes, columns):
    sns.histplot(data=df, x=column, ax=ax, kde=True, bins=30)  # Adjust bins for clarity
    ax.set_title(column, fontsize=12)
    ax.set_xlabel(column, fontsize=10)
    ax.set_ylabel('Frequency', fontsize=10)

# Hide unused subplot axes
for i in range(len(columns), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()  # Adjust layout for better readability
plt.show()
```

```python
desired_order = ['SURF_ppb_O3', 'SURF_ug_NO2', 'SURF_ug_PM25_rh50']  # Corrected column names

# Select only the specified features from the dataframe
selected_columns = df[desired_order]

# Recalculate the correlation matrix based on the selected features
corr_matrix = selected_columns.corr()

# Plot the heatmap
plt.figure(figsize=(13, 6))  # Set figure size

sns.heatmap(
    corr_matrix,                     # Correlation matrix with manually selected columns
    cmap='Blues',                     # Use 'Blues' color map
    annot=True,                        # Annotate with correlation values
    fmt='.2f',                         # Format annotation with 2 decimal places
    linewidths=0.5,                    # Add some space between cells
    vmin=-1, vmax=1,                   # Set color limits from -1 to 1 for correlation
    square=True,                       # Make the heatmap square
    cbar_kws={'shrink': 0.75}          # Resize the color bar for readability
)

plt.title("Correlation Heatmap of Selected Features", fontsize=14)
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import plotly.graph_objects as go
import plotly.subplots as sp

# Perform seasonal decomposition
result = seasonal_decompose(df['SURF_ppb_O3'], model='additive', period=365)

# Create subplots for the decomposed components
fig = sp.make_subplots(
    rows=4, cols=1, shared_xaxes=True,
    subplot_titles=['Observed', 'Trend', 'Seasonal', 'Residual']
)

# Add traces for each component
fig.add_trace(go.Scatter(x=df.index, y=result.observed, mode='lines', name='Observed'), row=1, col=1)
fig.add_trace(go.Scatter(x=df.index, y=result.trend, mode='lines', name='Trend'), row=2, col=1)
fig.add_trace(go.Scatter(x=df.index, y=result.seasonal, mode='lines', name='Seasonal'), row=3, col=1)
fig.add_trace(go.Scatter(x=df.index, y=result.resid, mode='lines', name='Residual'), row=4, col=1)

# Update layout with the correct template
fig.update_layout(
    template='plotly',
    height=800,
    title='O3',
    title_x=0.5
)

# Show the figure
fig.show()
```

```python
import netCDF4
import numpy as np
import matplotlib.pyplot as plt

# Open the NetCDF file
O3map = netCDF4.Dataset('BaseCase_PERT_O3_YEARLY.nc', 'r')

# Extract variables: lon, lat, and O3 concentration
lon = O3map.variables['lon'][:]
lat = O3map.variables['lat'][:]
O3 = O3map.variables['SURF_ppb_O3'][:]  # Assuming time is just one value

# Close the dataset after reading the necessary variables
O3map.close()

# Mask out invalid data (_FillValue) if necessary
fill_value = 9.969209968386869e+36
O3 = np.ma.masked_equal(O3, fill_value)

# Plot the O3 concentration for the single time point
plt.figure(figsize=(10, 6))
plt.contourf(lon, lat, O3[0, :, :], cmap='viridis')  # Time[0] means the first (and only) time point
plt.colorbar(label='O3 Concentration (g/m)')
plt.title('O3 Concentration Across the Region')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
O3map = netCDF4.Dataset('BaseCase_PERT_O3_YEARLY.nc', 'r')

# Extract the variables
lon = O3map.variables['lon'][:]
lat = O3map.variables['lat'][:]
time = O3map.variables['time'][:]
O3 = O3map.variables['SURF_ppb_O3'][:]  # Shape: (time, lat, lon)

# Close the dataset after reading the necessary variables
O3map.close()

# Flatten the 3D O3 data (time, lat, lon) into 2D
# Assuming there's only one time point here
time = np.repeat(time, lon.size * lat.size)  # Repeat the time for each lat/lon combination
lat_lon_pairs = np.array(np.meshgrid(lat, lon)).T.reshape(-1, 2)  # Create all combinations of lat/lon

# Create a DataFrame
df = pd.DataFrame(lat_lon_pairs, columns=['Latitude', 'Longitude'])
df['Time'] = time
df['O3 Concentration (g/m)'] = O3[0, :, :].flatten()  # Flatten the O3 array for the first time point

# Save to CSV
df.to_csv('O3_data.csv', index=False)

print("CSV file saved successfully!")
```
